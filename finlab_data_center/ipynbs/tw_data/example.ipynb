```python
df=pd.read_pickle(os.path.dirname(os.path.abspath("__file__"))+'/pickle_table/tw_data/stock_price.pkl')
```

```python
# df.to_pickle(os.path.dirname(os.path.abspath("__file__"))+'/pickle_table/tw_data/stock_price.pkl')
```

```python
df_cols = df.columns
cols = [i for i in df_cols if i not in ['stock_id', 'date', 'index','id']]
for c in cols:
    if str(df[c].dtype) in ['float64','int64']:
        print(c)
        pivot_df = pd.pivot_table(df, index=['date'], columns=['stock_id'], values=c,aggfunc='first')
        pivot_df.to_pickle(tw_fdata_path+f'/{c}.pkl')
```
